### Augmentation Parameters

```python
aug = augtools.Augmenter(pwa_scale=(0.05,0.075),
                         e_sev=3,
                         g_sev=2,
                         b_sev=2,
                         contrast=True,
                         clip_limit=(1,3),
                         dropout_pair=(0.15,0.85),
                         x_scale=(0.7,1.3),
                         y_scale=(0.7,1.05),
                         shear_range=(-20,20),
                         x_shift=(-0.05,0.05),
                         y_shift=(-0.1,0.1),
                         rotate_range=(-30,30),
                         v_flip=0.5,
                         h_flip=0.5,
                         sharpness=True,
                         fill_mode="reflect",
                         randomize_params=True)
```

### DS 2: Rebalanced Dataset

```python
%%time
ds_2 = uptools.Image_Dataset("data/img_data", "imgs", 256)
```

```python
%%time
ds_2.split_rebalance((0.76,0.12,0.12),
                     augmenter=aug,
                     augment_scale=700,
                     augment_type="simple",
                     order=[7,4,5,0,2,1,3,12,14,10,13,15,8,11,6,9])
```
